```python
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [25, 34, 30, 28, 38],
    "Salary": [50000, 70000, 60000, 55000, 60000],
    "Years_of_Experience": [2, 10, 4, 5, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.8, 4.0, 4.7, 3.5]
}

df = pd.DataFrame(data)
df
```

```python
shape = df.shape
print("shape of the dataframe is : " , shape)
```

```python
summary = df.dtypes
print(summary)
```

```python
stats = df.describe()
print(stats)
```

```python
print("first 5 rows : ")
print(df.head(5))
print("last 3 rows : ")
print(df.tail(3))
```

```python
print(df["Salary"].mean())
print(df["Bonus"].sum())
print(df["Age"].min())
print(df["Rating"].max())
```

```python
print(df.sort_values(by = "Salary" , ascending = False))
```

```python
def performance(rating):
  if rating >= 4.5 :
    return "Excellent"
  elif rating >= 4.0 and rating < 4.5:
    return "Good"
  else:
    return "Average"

df["Performance"] = df["Rating"].apply(performance)
df
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.rename(columns = {"Employee_ID" : "ID"} , inplace = True)
df
```
